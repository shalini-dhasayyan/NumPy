```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_points = np.arange(-8,8,0.04)
```

```python
dx,dy=np.meshgrid(x_points,x_points)
```

```python
dx
```

```python
dy
```

```python
z=(np.sin(dx)+np.sin(dy))
```

```python
z
```

```python
plt.imshow(z,cmap='Dark2')
```

```python
plt.imshow(z,cmap='tab10')
plt.colorbar()
plt.title('sin(dx)+sin(dy)')
```

```python
X=np.array([2,5,8,10])
Y=np.array([100,200,300,400])
```

```python
condition=np.array([True,True,False,False])
```

```python
answer1=[(X_val if cond else Y_val) for X_val,Y_val,cond in zip(X,Y,condition)]
```

```python
answer1
```

```python
answer2=np.where(condition,X,Y)
```

```python
answer2
```
